```python
import pandas as pd
import os
import fonctions.fonctions as f
from euroleague_api.play_by_play_data import PlayByPlay
```

```python
data_dir = r"C:\Users\guill\OneDrive\Documents\euroleague_eurocup\datas"
competition = "euroleague"
season = 2025
```

```python
competition_pbp = pd.read_csv(os.path.join(data_dir, f'{competition}_pbp_{season}.csv'))
competition_gs = pd.read_csv(os.path.join(data_dir, f'{competition}_gs_{season}.csv'))

game_code_list = competition_gs["Gamecode"].unique().tolist()
```

```python
df = f.calcul_per2(data_dir, season, competition)
df = df[(df["TIME_ON"]>0)].copy()
df.insert(6, "I_PER", df["PER"] / df["TIME_ON"] ** 0.5)
df["I_PER"] = df["I_PER"].round(2)
df["NB_GAME"] = 1
df = df[['ROUND', 'NB_GAME', 'TEAM', 'OPPONENT', 'HOME', 'WIN', 'NUMBER', 'PLAYER',
         'TIME_ON', "I_PER", 'PER', 'PM_ON', 'PTS', 'DR', 'OR', 'TR', 'AS', 'ST', 'CO',
         '1_R', '1_T', '2_R', '2_T', '3_R', '3_T','TO', 'FP', 'CF', 'NCF']]
```

```python
data = f.get_aggregated_data(df, 1, 38)
```

```python
data["PIR"] = (
    data[["PTS","TR","AS","ST","CO","1_R","2_R","3_R","FP"]].sum(axis=1)
    - data[["1_T","2_T","3_T","TO","CF","NCF"]].sum(axis=1)
)

data["SCORE_FANTASY"] = data["PIR"] + (
    (data["WIN"] == "YES") * data["PIR"].abs() * 0.10
)

```

```python
df_player = data.groupby("PLAYER").agg(
    TOTAL_SCORE_FANTASY=("SCORE_FANTASY", "sum"),
    N_MATCH=("PLAYER", "count")
).reset_index()
```

```python
df_player.sort_values(by="TOTAL_SCORE_FANTASY", ascending=False, inplace=True)
df_player.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
import os
import fonctions.fonctions as f
```

```python
data_dir = r"C:\Users\guill\OneDrive\Documents\euroleague_eurocup\datas"
competition = "euroleague"
season = 2025
```

```python
competition_gs = pd.read_csv(os.path.join(data_dir, f'{competition}_gs_{season}.csv'))
```

```python
competition_gs.columns
```

```python
cgs = competition_gs[["local.club.abbreviatedName","road.club.abbreviatedName","local.score","road.score"]]
cgs.columns = ["HOME_TEAM","AWAY_TEAM","HOME_SCORE","AWAY_SCORE"]
```

```python
# 1) Calcul de la colonne résultat (1 si home win, sinon 0)
cgs["resultat"] = (cgs["HOME_SCORE"] > cgs["AWAY_SCORE"]).astype(int)

# 2) Liste de tous les clubs
clubs = pd.unique(cgs[["HOME_TEAM", "AWAY_TEAM"]].values.ravel())

# 3) Création d’un df vide avec 0 partout
club_matrix = pd.DataFrame(0, index=cgs.index, columns=clubs)

# 4) Remplissage : +1 pour home, -1 pour away
for idx, row in cgs.iterrows():
    club_matrix.at[idx, row["HOME_TEAM"]] = 1
    club_matrix.at[idx, row["AWAY_TEAM"]] = -1

# 5) Concaténation avec le résultat
cgs_transformed = pd.concat([club_matrix, cgs["resultat"]], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

# X = colonnes clubs
X = cgs_transformed.drop(columns=["resultat"])
y = cgs_transformed["resultat"]

# Modèle de régression logistique
logreg = LogisticRegression(max_iter=500)
logreg.fit(X, y)

# Récupération des coefficients
coefs = pd.Series(logreg.coef_[0], index=X.columns)

# Classement des coefficients
coefs_sorted = coefs.sort_values(ascending=False)

# Classement basé sur les coefficients triés (déjà dans coefs_sorted)
ranking = coefs_sorted.rank(ascending=False, method="first")

# Conversion en tableau propre
classement = pd.DataFrame({
    "coef": coefs_sorted,
    "rang": ranking.astype(int)
}).sort_values("rang")

print(classement)
```
